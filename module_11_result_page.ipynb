## 2. Model Accuracy Comparision

### 2.1 Abstract catagory Based Sentiment Analysis

```python
from prettytable import PrettyTable

#If you get a ModuleNotFoundError error , install prettytable using: pip3 install prettytable

x = PrettyTable()
x.field_names = ["Dataset", "Xgboost:train_acc", "Xgboost: test_acc", " CNN:train_acc","CNN:test_acc"]

x.add_row(["Restaurant Data" ,73,65,99,70])
x.add_row(["Google bert augmented", 75,71,100,91])
x.add_row(["RTT augmented", 75,72,99,91])
x.add_row(["Restaurant_Data + Google_bert_aug + RTT_aug" , 75,73,100,99])
print(x)
```

### 2.2 Polarity Based Sentiment Analysis

```python
x = PrettyTable()
x.field_names = ["Dataset", "Xgboost:train_acc", "Xgboost: test_acc", " CNN:train_acc","CNN:test_acc"]

x.add_row(["Restaurant Data" ,68,58,98,63])
x.add_row(["Google bert augmented", 69,69,99,90])
x.add_row(["RTT augmented", 69,65,99,87])
x.add_row(["Restaurant_Data + Google_bert_aug + RTT_aug" , 68,66,100,99])
print(x)
```

## Observation :
       Both the models , Abstract catagory Based Sentiment Analysis and  Polarity Based Sentiment Analysis model accuracies comparision shows that the cnn model works well on all three types of data, Resturant data, bert augmented and RTT augmented than the xgbbost. CNN accuracies are better than xgbbost.
